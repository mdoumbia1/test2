```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)

# Show the plot
plt.show()
```

```python
sns.barplot(x="day", y="tip", hue="sex", data=tips)

# Display the plot
plt.title('Average Tip Amount by Day with Gender Hue')
plt.show()
```

```python
# Scatter plot with regression line
sns.lmplot(x="total_bill", y="tip", data=tips)

# Display the plot
plt.title('Scatter Plot of Total Bill vs. Tip with Regression Line')
plt.show()
```

```python
# Joint plot with KDE
sns.jointplot(x="total_bill", y="tip", data=tips, color='red',kind="kde", fill=True)

# Display the plot
plt.suptitle('Joint Distribution of Total Bill and Tip with KDE', y=1.02)
plt.show()
```
